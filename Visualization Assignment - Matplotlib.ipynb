Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer: Matplotlib is a python library used for data visualization. We can use it for data analysis and creating representation of our data. We can customize the visualizations as per our requirements and can also add text to explain the x and y axis.

It is majorly used for data analysis purposes through visualization and also helps in presenting the data beautifully.

5 plots that can be plotted using the pyplot module of matplotlib:
1. Scatter
2. bar
3. Histogram(hist)
4. line plot(plot)
5. Piechart

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# Answer to Question 2
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my title")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
```

```python
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

Answer: Subplot function is used to plot multiple figures within a single figures.  Generally, when we have to compare different datasets of same type, subplot function is very useful as we can visualize different dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Answer: Bar plot is a commonly used visualization where the data is represented in the horizontal or vertical both. It is majorly used to plot a numerical data against categorial data as shown in the visualization below. Company names like Apple, microsoft etc are categorial data presenting the numerical value such as their profits.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Answer: Box plot is a quartile based data representation. It resemples the shape of our data. It presents the data through the median, upper and lower quartiles, minimum and maximum values and any outliers in the dataset. 

Box plot is majorly used for comparison purposes between multple groups. We see this kind of data a lot in the share price analysis as well.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
Data = [box1, box2]
```

```python
plt.boxplot(Data)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('presenting the boxplot')
plt.show()
```
